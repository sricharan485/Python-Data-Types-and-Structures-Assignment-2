Syntax: Lists are defined using square brackets [], while tuples are defined using parentheses ().

10. Explain why strings are immutable in Python.

Answer:
  
 - Strings are immutable in Python because this design choice has several advantages:

 - Efficiency: Immutability allows for certain optimizations, such as sharing the same string object in memory when multiple variables point to the same string value.
  
 - Thread Safety: Immutable objects are inherently thread-safe because their state cannot be changed by multiple threads simultaneously.
  
 - Hashability: Immutable objects can be used as keys in dictionaries and     elements in sets because their hash value remains constant.

  Represent relationships between data: Dictionaries are ideal for representing objects or structures with named attributes.

14. How does the “in” keyword work differently for lists and dictionaries?

Answer:

- For lists, the in keyword checks if an element is present among the values    in  the list. It performs a linear search, so the time complexity is O(n).
 For dictionaries, the in keyword checks if a key exists in the dictionary. It uses hashing for efficient lookup, so the average-case time complexity is O(1).

17. Describe the time complexity of accessing elements in a dictionary.

Answer:

 - The average-case time complexity of accessing elements in a dictionary using its key is O(1) (constant time). This is because dictionaries use hash tables for efficient key lookups. In the worst case (due to hash collisions), it can be O(n), but this is rare with good hash functions.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?

Answer:

 - Historically, dictionaries in Python were considered unordered because the order of elements was not guaranteed to be the insertion order. However, since Python 3.7, dictionaries maintain insertion order. While this makes them ordered in practice, the fundamental way they store data (using hash tables) means that accessing elements is based on the key's hash value, not its position in the dictionary. This doesn't affect data retrieval speed (still O(1) on average), but it's a conceptual difference from sequence types like lists where order is inherent to access.
